# Advance Operation with Business Examples

```python
import numpy as np;
import matplotlib.pyplot as plt
```

```python
# Data Structure: [restaurante_id,2021,2022,2023,2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])
print("Zomato sles anaylysis")
print("\nsales data shape: \n",sales_data.shape)

print("\nsample data for first 3 restaurant\n",sales_data[:3])
```

```python
#total sales per year
print(np.sum(sales_data,axis=0));
yearlyTotal = np.sum(sales_data[:,1:],axis=0);
print("\n",yearlyTotal)
```

### Minimum sales per restaurant

```python
minimum_sales = np.min(sales_data[:,1:],axis=1)
print(minimum_sales)
```

### Maximum sales per year|

```python
max_sales = np.max(sales_data[:,1:],axis=0)
print("\nmax sales: \n",max_sales)
```

### Average sales per restaurant

```python
avg_salary = np.mean(sales_data[:,1:],axis=1)
print("\n avg salary per restaurant: \n",avg_salary);
```

### get the cumalative sales means first and then second + first then third+second+first

```python
cumalative_sales = np.cumsum(sales_data[:,1:],axis=1)
print("\ncumalative Sales is: \n",cumalative_sales)
```

```python
plt.Figure(figsize=(10,6))
plt.plot(np.mean(cumalative_sales,axis=0))
plt.title("Avg cumalative sales accross all restaurant: ")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```
